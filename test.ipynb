```python
from pqcrypto.kem.ml_kem_512 import generate_keypair, encrypt, decrypt

# Step 1: Server generates public and secret key
public_key, secret_key = generate_keypair()

print("Public key generated")
print("Secret key generated")

# Step 2: IoT device encrypts using server public key
ciphertext, shared_secret_sender = encrypt(public_key)

print("IoT device created encrypted secret")

# Step 3: Server decrypts ciphertext
shared_secret_receiver = decrypt(secret_key, ciphertext)

print("Server recovered shared secret")

# Step 4: Verify shared secrets match
print("Shared secret match:", shared_secret_sender == shared_secret_receiver)
```

```python
from pqcrypto.kem.ml_kem_512 import generate_keypair, encrypt, decrypt
import hashlib

# Generate keypair
public_key, secret_key = generate_keypair()

# IoT device encrypts
ciphertext, shared_secret_sender = encrypt(public_key)

# Server decrypts
shared_secret_receiver = decrypt(secret_key, ciphertext)

# Convert shared secret to symmetric key
key = hashlib.sha256(shared_secret_receiver).digest()

print("Derived symmetric key:", key[:16])
```

```python
import time
from pqcrypto.kem.ml_kem_512 import generate_keypair, encrypt, decrypt

# Key generation benchmark
start = time.time()
public_key, secret_key = generate_keypair()
keygen_time = time.time() - start

# Encryption benchmark
start = time.time()
ciphertext, ss = encrypt(public_key)
enc_time = time.time() - start

# Decryption benchmark
start = time.time()
decrypt(secret_key, ciphertext)
dec_time = time.time() - start

print("Key generation time:", keygen_time)
print("Encapsulation time:", enc_time)
print("Decapsulation time:", dec_time)
```

```python
import time
from pqcrypto.kem.ml_kem_512 import generate_keypair, encrypt, decrypt

iterations = 100

start = time.time()
for _ in range(iterations):
    public_key, secret_key = generate_keypair()
print("Average keygen:", (time.time() - start)/iterations)

start = time.time()
for _ in range(iterations):
    ciphertext, ss = encrypt(public_key)
print("Average encrypt:", (time.time() - start)/iterations)

start = time.time()
for _ in range(iterations):
    decrypt(secret_key, ciphertext)
print("Average decrypt:", (time.time() - start)/iterations)
```
